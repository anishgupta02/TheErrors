```python
(b,g,r)=cv2.split(img)
```

```python
b.shape,g.shape,r.shape
```

```python
cv2_imshow(b),cv2_imshow(g),cv2_imshow(r)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.subplot(1,3,1)
plt.imshow(b,cmap="Greys")
plt.subplot(1,3,2)
plt.imshow(g,cmap="Greys")
plt.subplot(1,3,3)
plt.imshow(r,cmap="Greys")
plt.show()
```

```python
cv2_imshow(cv2.merge([b,g,r]))
```

```python
cv2_imshow(cv2.blur(img,(10,10)))
```

```python
cv2_imshow(cv2.medianBlur(img,7))
```

```python
cv2_imshow(cv2.GaussianBlur(img,(3,3),6))
```

```python
cv2_imshow(cv2.bilateralFilter(img,70,90,90))
```

```python
g=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(cv2.threshold(g,150,180,cv2.THRESH_BINARY)[1])
```

```python
edges=cv2.Canny(img,100,250)
```

```python
cv2_imshow(edges)
```

```python
import numpy as np
d=cv2.dilate(edges,np.ones((3,3),dtype=np.uint8))
cv2_imshow(d)
```

```python
cv2_imshow(cv2.erode(d,np.ones((3,3),dtype=np.uint8)))
```
